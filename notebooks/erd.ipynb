# Entity Relationship Diagram for the Config pydantic model

The Config model is a hierarchical data model. In this notebook, we plot the ERD for the data model
using the erdantic Python library. The generated ERD uses Crow's Foot notation; in this model,
all entity relationships have a cardinality of one, represented by the double-dash.

```python
import erdantic
from histopath_bim_des.config import Config

erdantic.draw(Config, out='Config.erd.png')
from IPython.display import Image
Image('Config.erd.png', width=1024)
```
